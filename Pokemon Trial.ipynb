### Making changes to the data

```python
# Adding new columns
df['Total'] = df['HP']+ df['Attack']+df['Defense']+df['Sp. Def']+df['Speed']+df['Sp. Atk']

# Another way to add columns is using the iloc function. (axis = 1 is used to add column horizontally, asix=0 used vertically)
df['Total'] = df.iloc[:,4:10].sum(axis=1)
df.head(5)
```

```python
# Removing a column
df = df.drop(columns =['Total'])
df.head(5)
```

```python
# Changing the position of a colunm
cols=list(df.columns.values)
df=df[cols[0:4] +[cols[-1]]+cols [4:12]]
df.head(5)
```

### Filtering data

```python
# df.head(49)
# Filtering by specific column
# df.loc[df['Type 2'] == 'Poison']

# filtering by multiple column. & rep AND | rep OR > rep GREATER THAN < rep LESS THAN ~ rep NOT 
df.loc[(df['Type 2'] == 'Poison') & (df['Type 1'] == 'Grass') & (df['HP'] > 70)]
```

```python
# saved above data as it is
new_df = df.loc[(df['Type 2'] == 'Poison') & (df['Type 1'] == 'Grass') & (df['HP'] > 70)]
new_df.to_csv('Filtered2.csv',index = False)
new_df = new_df.reset_index()
```

```python
# the index changes here after reset_index from 2,3,50 to 0,1,2 
new_df
```

```python
# to get rid of the old index totally we use
new_df.reset_index(drop=True)
new_df
```

```python
# Filtering by locating a specific word in a colunm. ~ rep NOT
# df.loc[df['Name'].str.contains('Mega')]
df.loc[~df['Name'].str.contains('Mega')]
```

```python
# Using REGEX
# flags=re.I- used to ignore cases of words
import re
df.loc[df['Type 1'].str.contains('Grass|fire',flags=re.I,regex=True)]
```

```python
# all words in name column begining with Pi. the * means zero
df.loc[df['Name'].str.contains('^Pi[a-z]*',flags=re.I,regex=True)]
```

### AGGREGATE STATISTICS(GROUPBY)

```python
# data reloaded without the test options
df=pd.read_csv('Updated.csv')
df
```

```python
# finding the mean of type 1
# you can find sum,mean and count using groupby
df.groupby(['Type 1']).mean().sort_values('Attack', ascending =False)
```

```python
df.groupby(['Type 1']).sum()
```

```python
df['count']=1
df.groupby(['Type 1']).count()
```

```python
# or easier way to find the count 
df.groupby(['Type 1']).count()['count']
```
